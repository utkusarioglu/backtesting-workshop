```python
import pandas as pd
import numpy as np
```

```python
import sys
if sys.modules.get("optimized_ma") is not None: 
    del sys.modules["optimized_ma"]
from optimized_ma import OptimizedMa

PAIRS = ["BTC", "ETH"]

o = OptimizedMa("USD", PAIRS, "temp_data.json")
o.get_data(3, "1m")
o.create_indicator()
pf = o.run(
    rsi_window=np.arange(10, 50, step=5, dtype=int),
    rsi_period=["5T", "10T"],
    ma_window=np.arange(20, 100, step=10, dtype=int),
    entry_threshold=np.arange(30, 40, step=2, dtype=int),
    exit_threshold=np.arange(60, 70, step=2, dtype=int),
)
```

```python
pf.trades.close
```
