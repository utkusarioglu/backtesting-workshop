```python
PAIRS = ["BTC", "ETH"]
o = OptimizedMa("USD", PAIRS, "temp_data.json")
prices = o.get_data(3, "1m")
o.create_indicator()
pf = o.run(
    rsi_window=np.arange(10, 50, step=5, dtype=int),
    rsi_period=["5T", "10T"],
    ma_window=np.arange(20, 100, step=10, dtype=int),
    entry_threshold=np.arange(30, 40, step=2, dtype=int),
    exit_threshold=np.arange(60, 70, step=2, dtype=int),
)
```

```python
def plot_btc_usd_price_response():
    btc_usd_er = prices["BTC-USD"]
    eth_usd_er = prices["ETH-USD"]
    eth_btc_er = btc_usd_er[0] / eth_usd_er[0]
    fig = btc_usd_er.vbt.plot()
    (eth_usd_er * eth_btc_er).vbt.plot(fig=fig)
    fig.show()

plot_btc_usd_price_response()
```

```python
def plot_btc_signals():
    btc_usd_er = prices["BTC-USD"]
    fig = btc_usd_er.vbt.plot()
    max_index = pf.total_return().idxmax()
    entries = o.entries[max_index]
    exits = o.exits[max_index]
    fig = entries.vbt.signals.plot_as_entry_markers(btc_usd_er, fig=fig)
    fig = exits.vbt.signals.plot_as_exit_markers(btc_usd_er, fig=fig)
    fig.show()
plot_btc_signals()
```
