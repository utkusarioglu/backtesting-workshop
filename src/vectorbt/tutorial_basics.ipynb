```python
import vectorbt as vbt
from datetime import datetime, timedelta
```

```python
pairs = ["BTC-USD", "ETH-USD"]
crypto_close = vbt.YFData.download(pairs[0], missing_index="drop").get("Close")
```

```python
crypto_close
```

```python
rsi = vbt.RSI.run(crypto_close, window=14)
```

```python
rsi.rsi
```

```python
entries = rsi.rsi_crossed_below(30)
exits = rsi.rsi_crossed_above(70)
```

```python
pf = vbt.Portfolio.from_signals(crypto_close, entries, exits)
pf
```

```python
pf.stats()
```

```python
pf.total_return()
```

```python
pf.plot()
```

```python
end_date = datetime.now()
start_date = end_date - timedelta(days=3)
cryptos = "BTC ETH XMR ADA"
pairs = [f"{c}-USD" for c in cryptos.split(" ")]
crypto_close = vbt.YFData.download(
    pairs, missing_index="drop", interval="1m", start=start_date, end=end_date
).get("Close")
crypto_close
```

```python
windows = [7, 14, 21]
rsi = vbt.RSI.run(crypto_close, window=windows)
entries = rsi.rsi_crossed_below(30)
exits = rsi.rsi_crossed_above(30)
pf = vbt.Portfolio.from_signals(crypto_close, entries, exits)
pf
```

```python
pf.stats()
```

```python
pf.total_return()
```
